```python
import pandas as pd
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
data = pd.read_excel('height.xls')
data.sample(5)
```

```python
data.plot(kind='scatter', x='height', y='weight', figsize=(10, 10), c='gender')
```

```python
data.plot(kind='scatter', x='height', y='weight', c=data.gender, figsize=(10, 10))
```

```python
males = data.loc[data.gender==0, 'height':'weight']
females = data.loc[data.gender==1, 'height':'weight']
```

```python
ax = males.plot(kind='scatter', x='height', y='weight', label='male')
```

```python
females.plot(kind='scatter', x='height', y='weight', label='female', ax=ax)
```
