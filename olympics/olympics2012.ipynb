```python
import pandas as pd
import seaborn as sns
import numpy as np

%matplotlib inline
sns.set()
```

```python
data = pd.read_excel('olympics_2012.xls')
data.head()
```

```python
sns.lmplot(x='Year', y='Time', data=data, fit_reg=True, size=8)
```

```python
data.corr()
```

#### Least squares regression

line equation, the first number is the slope the second is the

```python
least_squares_line = np.polyfit(data.Year, data.Time, 1)
least_squares_line
```

### Without outlier

```python
data.shape
```

```python
no_outlier = data[~(data.Time > 250)]
```

```python
no_outlier.shape
```

```python
sns.lmplot(x='Year', y='Time', data=no_outlier, fit_reg=True, size=8)
```

#### Least squares regression

```python
np.polyfit(no_outlier.Year, no_outlier.Time, 1)
```
