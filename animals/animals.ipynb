```python
import pandas as pd
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
data = pd.read_excel('animals.xls')
data.head()
```

### Correlation

```python
data.corr()
```

### Correlation withou elephant

```python
data[data.animal != 'elephant'].corr()
```

```python
data.plot(kind='scatter', x='longevity', y='gestation', figsize=(10, 10))
```

```python
sns.lmplot(data=data, x='longevity', y='gestation', size=10, fit_reg=True)
```
